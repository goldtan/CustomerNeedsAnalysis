# Data preprocessing

```python
import pandas as pd
```

```python
review_all = pd.read_json("AMAZON_FASHION.json", orient = 'columns', lines = True)
review_all
```

```python
review_all.rename(columns = {'asin' : 'productID', 'overall' : 'rating'}, inplace = True)
```

```python
#3점 미만의 리뷰만 & 다른 사람에게 공감을 한 개 이상 받은 리뷰만 
review_neg = review_all[(review_all['rating'] < 3) & (review_all['vote'] > 0)] 
```

```python
review_neg.columns
```

```python
review_neg = review_neg.drop(columns = ['verified', 'reviewTime', 'reviewerID', 'reviewerName', 'unixReviewTime', 'vote', 'image'])
review_neg = review_neg.reset_index(drop = True)
review_neg
```

```python
review_neg['reviewTotal'] = review_neg['reviewText'] + " " + review_neg['summary']
data = review_neg
data
```

# Text preprocessing

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag
import re
```

```python
review_total = data['reviewTotal'].map(lambda x: str(x).lower()) #소문자로
review_total
```

```python
review_total = review_total.map(lambda x: re.sub('[^a-zA-Z]', ' ', x)) #특수문자와 숫자 제거
review_total
```

```python
review_total = review_total.map(word_tokenize) #토큰화
review_total
```

```python
review_total = review_total.map(pos_tag)
review_total
```

```python
#형용사만 사용했을 때가 가장 좋은 결과
nouns = ['NN', 'NNP', 'NNS', 'NNPS']
verbs = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
adverbs = ['RB', 'RBR', 'RBS']
adjective = ['JJ', 'JJR', 'JJS']
```

```python
lemmatizer = WordNetLemmatizer() 

#표제어 추출 함수
def lemmatize_word(x):
    a = []
    for tup in x: 
        if tup[1] in adjective :
            a.append(lemmatizer.lemmatize(tup[0], 'a'))
        elif tup[1] in nouns:
            a.append(lemmatizer.lemmatize(tup[0], 'n'))
        else :
            pass 
    return a
```

```python
review_total = review_total.map(lemmatize_word) #표제어 추출
review_total
```

```python
stoplist = set(stopwords.words('english'))

#불용어 제거 함수
def stop_word(x):
    a = []
    for word in x: 
        if word not in stoplist: 
            a.append(word)
    return a
```

```python
review_total = review_total.map(stop_word) #불용어 제거
review_total
```

```python
data['reviewWord'] = review_total
data.to_csv("data_pre.csv")
data.to_pickle("data_pre.pkl")
data
```
